### Import the relevant libraries

```python
import numpy as np
```

### Generate random input data to train on

```python
from numpy.core.fromnumeric import shape
obs_size = 1000
x = np.random.uniform(-10,10, size=(obs_size,1))
z = np.random.uniform(-5,5, size=(obs_size,1))

inputs = np.concatenate([x,z], axis=1)
```

### create target

```python
noise = np.random.uniform(-1,1, size=(1000,1))
target = 5*x + 3*z + 15 + noise
```

### create random weights and baises

```python
init_range = 0.1
weights = np.random.uniform(-init_range,init_range, size=(2,1))
baises = np.random.uniform(-init_range,init_range, size=1)
```

```python
 inputs.shape,weights.shape,
```

### building a model

```python
outputs = np.dot(inputs,weights) + baises
outputs.shape
```

```python
baises.shape
```

```python
weights.shape
```

```python
lr = 0.01
for i in range(10000):
  outputs = np.dot(inputs,weights) + baises

  delta = outputs-target
  delta_scaled = delta/obs_size
  loss = np.sum(delta ** 2) / 2 / obs_size

  print(loss)

  #optimization 
  weights = weights - lr * np.dot(inputs.transpose(),delta_scaled)
  baises = baises - lr * np.sum(np.dot(inputs.transpose(),delta_scaled))

```
