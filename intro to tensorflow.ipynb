### generate random data

```python
#import libraries
import numpy as np
import tensorflow as tf
```

```python
observations = 1000
xs = np.random.uniform(low=-10, high=10, size=(observations,1))
zs = np.random.uniform(-10, 10, (observations,1))
noise = np.random.uniform(-1, 1, (observations,1))
targets = 2*xs - 3*zs + 5 + noise

inputs = np.concatenate([xs,zs] , axis= 1)
np.savez("tf_intro", inputs, targets)
```

### solving with tf

```python
train = np.load("tf_intro.npz")
```

```python
train["inputs"]
```

```python
from tensorflow.python import metrics
model= tf.keras.Sequential([
    tf.keras.layers.Dense(1)
])

model.compile(
    loss = "mse",
    optimizer = "sgd",
)

history = model.fit(inputs,targets, epochs = 100)
```

```python
model.get_weights()
```

### plot outputs against the targets

```python
import matplotlib.pyplot as plt
pred = model.predict(inputs)
plt.scatter(pred , targets, alpha =0.2)
plt.title("Evaluation" )

```
